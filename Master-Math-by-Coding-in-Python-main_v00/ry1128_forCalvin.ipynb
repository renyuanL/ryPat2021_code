```python
#(1): y= x1**2 + x2**2
#(2): v= w1*x1 + w2*x2 + b
# 根據 (1) 產生 1000點 資料
# 根據 (2) 用 v 來近似 y
# 要找出最佳的 w1, w2, b
# 使得 代價函數(cost function) J= Σ [(y-v)**2] for all 1000 點資料 為最小 
```

```python
#[1] 先 製造 1000 點 資料

import numpy as np

np.random.seed(0)
X= np.random.random(size= (1000,2))
X
```

```python
x1= X[:,0]
x2= X[:,1]

y= x1**2 + x2**2
y
```

```python
np.random.seed(1)
w= np.random.random(size=(3,1))
w
```

```python
w1, w2, b= w
v= w1*x1+ w2*x2 + b
v
```

```python
J= ((v-y)**2).sum()
J
```

```python
for i in range(1000):
    
    dJ_dw1= (2*(v-y)*x1).sum()  # 這是 gradient (梯度 or 微分)
    dJ_dw2= (2*(v-y)*x2).sum()
    dJ_db=  (2*(v-y)*1).sum()

    η=  1e-5 # 這是 learning rate

    δw1= -dJ_dw1 * η
    δw2= -dJ_dw2 * η
    δb=  -dJ_db  * η

    w1 += δw1  # 這是 Newton 疊代公式，會讓 J 逐漸變小
    w2 += δw2
    b  += δb

    v= w1*x1+ w2*x2 + b
    v
    J= ((v-y)**2).sum()  # 檢查 J 是否真的逐漸變小
    J
J  # 檢查最後的 J
```
