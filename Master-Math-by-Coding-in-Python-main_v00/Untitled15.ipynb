```python
import numpy as np
import matplotlib.pyplot as plt

# pdensities of x, x^2, and x^3
%matplotlib qt
t= np.linspace(-10,10,1000)
fig, ax= plt.subplots(2,5)
fig.set_size_inches(12,6) # width, height in inches
for i in range(5):
    #i=1    
    # histogram of the function
    values= t**i
    
    #np.random.shuffle(values) # 加上這行頗為有趣
    
    counts, value_bins= np.histogram(values, 30)
    counts= counts #/ sum(counts)
    
    # plot the function
    ax[0,i].plot(values, t)
    ax[0,i].set_xlabel('values')
    ax[0,i].set_ylabel('t')
    ax[0,i].set_title(f'$values= t^{i}$')
    
    # plot its distribution
    ax[1,i].stem(
        value_bins[:-1],
        counts)
    ax[1,i].set_xlabel('value_bins')
    ax[1,i].set_ylabel('counts')
    
    
plt.show()
```

```python
#
# 這圖超有梗: 
# 可以再說明一下...
#
# sort(), shuffle(), histogram(), cumsum(), diff(), modf()
#
#
t= np.linspace(0,10,100)
x= t.copy()
y= abs(np.modf(x/10)[0]-.5)*2  #有梗
#y= y**2  #可以變化試試看
y1= y.copy()
y1= np.sort(y1) #有梗
y2= y.copy()
np.random.shuffle(y2) #有梗
ax= pl.axes()
ax.clear()
ax.plot(x, y,  'ro-')
ax.plot(x, y1, 'g^--')
ax.plot(x, y2, 'bs--')
f, x1= np.histogram(y) #有梗
x1= x1[:-1]            #有小梗
ax.barh(x1, width=f/10, 
        height=.06, color='magenta', alpha= .5, align='edge')
F= f.cumsum()  #有梗
ax.barh(x1, width=F/10, 
        height=.08, color='gray', alpha= .5, align='edge')
dF= np.diff(F) #有梗
dF= np.insert(dF,0,F[0]) # 這裡也有小小的梗
ax.barh(x1, width= dF/10, 
        height=.02, color='purple', alpha= .5, align='edge')
ax.set_title('y= y(t), f= f(y), F= F(y), dF= dF(y)')
ax.set_xlabel('t, f, F')
ax.set_ylabel('y')
ax.legend(['y= y(t)','y1= sort(y)','y2= shuffle(y)',
           'f= histogram(y)','F= cumsum(f)', 'dF= diff(F)'])
```

```python
# 中央極限定理 Central Limit Theorem
#
# generate data from various distributions, then sum them
# the sum will approach a normal distribution
#
sampleSize= 1000
t= np.arange(sampleSize)  
# 加總 100 個 隨機變數，陽春Python經典作法
X= 0 
for i in range(100):
    X += np.random.random(size= sampleSize)
# 加總 100 個 隨機變數， Python 經典做法 之2: list comprehension 
X100= [np.random.random(sampleSize)  for i in range(100)]
X=    sum(X100)
# 加總 100 個 隨機變數， numpy 經典做法，要注意 shape，axis
X100= np.random.random((100, sampleSize)) # .shape= (100, sampleSize)
X=    X100.sum(axis=0) # .shape= (1, sampleSize) ==> 再進化為 (sampleSize, )
f, x= np.histogram(X)
x= x[:-1]
fg, ax= pl.subplots(1,2)
for i in range(0,100,2):
    ax[0].plot(t, X100[i], alpha= .1)
ax[0].plot(t,X)
ax[1].plot(f,x)
ax[0].set_xlabel('t')
ax[0].set_ylabel('X')
ax[0].set_title('sum 100 independent random variables')
ax[1].set_xlabel('f')
ax[1].set_ylabel('x')
ax[1].set_title(' the sum of them will approach a Normal distribution')
ax[0].set_ylim([0,100])
ax[1].set_ylim([0,100])
```

```python
import numpy  as np
import pandas as pd
%matplotlib qt
pl.rcParams['axes.prop_cycle']= pl.cycler(
    color=['red', 
           'orange', 
           'yellow', 
           'green',
           'cyan',
           'blue',
           'purple',
           'magenta',
           'gray',
           'black'])
    
    
def ryUniformVsNormal(sampSize= 1000):
    
    X= pd.DataFrame(np.zeros((sampSize, 1)))
    for i in range(0,8):
        a= i-1/2
        b= i+1/2
        X[i]= np.random.uniform(
                low=  a, 
                high= b,
                size= sampSize)
    
    #X= X.cumsum()
    X.plot(title= 'uniform', alpha= .5)
    X.hist()
    
    
    Y= pd.DataFrame(np.zeros((sampSize, 1)))
    for i in range(0,8):
        μ= i
        σ= 1/2
        Y[i]= np.random.normal( 
                loc=   μ,  
                scale= σ,
                size= sampSize)
    #Y=Y.cumsum()
    Y.plot(alpha= .5, title= 'normal')
    Y.hist()
    
ryUniformVsNormal(sampSize= 1000)
```

```python
import numpy as np
import scipy.stats as st
xx= np.linspace(0.2,10,1001)
pl.figure()
θ= 2
for ν in range(1,10):
    a= ν/2
    yy= st.gamma(a= a , scale= θ).pdf(xx)
    pl.plot(xx,yy, label=f'a= {a}, θ= {θ}')
pl.title('Γ(a, θ).pdf(x)')
pl.xlabel('x')
pl.legend()
pl.figure()
for ν in range(1,10):
    yy= st.chi2(df= ν).pdf(xx)
    pl.plot(xx,yy, label=f'ν= {ν}')
pl.title('$χ^2(ν).pdf(x)$')
pl.xlabel('x')
pl.legend()
```
