```python
# import data with pandas, name variables x and y, determine (visually) how many clusters are in the data
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
data= pd.read_csv('data.csv',header=None,names=['x','y'])
data
```

```python
%matplotlib qt
sns.set_theme()

sns.jointplot(
    x= data['x'],
    y= data['y'],
    kind='kde',
    levels=20,
    cmap= 'rainbow'
).plot_joint(sns.scatterplot)

plt.show()
```

```python
penguins= sns.load_dataset("penguins")
q= sns.jointplot(
    data= penguins, 
    x=    "flipper_length_mm", 
    y=    "bill_length_mm", 
    hue=  "species",
    cmap= 'rainbow'
)
```
