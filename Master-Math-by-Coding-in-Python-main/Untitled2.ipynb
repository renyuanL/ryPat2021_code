```python
def f(x):
    y= x**2
    return y

π= sm.pi

p= sm.plotting.plot(
    f(x), 
    f(x).diff(x),
    f(x).integrate(x),
    (x,-1, 1), 
    show=False
)

p[0].line_color= 'r'
p[1].line_color= 'g'
p[2].line_color= 'b'
p.legend= True
p.show()
```

```python
def f(x):
    y= x**2
    return y

F= lambda x: f(x).integrate(x)
ϕ= lambda x: f(x).diff(x)

f(x), F(x), ϕ(x)
```
