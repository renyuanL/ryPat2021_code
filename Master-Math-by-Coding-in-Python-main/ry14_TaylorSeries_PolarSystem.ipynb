```python
## Finding Taylor polynomial

a= 1
f= sm.sin(x)

n= 5

T=   0 #f.subs(x,a) 
for i in range(0, n):
    T += f.diff(x,i).subs(x,a) *(x-a)**i /sm.factorial(i)   
T.expand()
```

```python
## plotting f(x) and T(x) in same window


p= sm.plot(f, T, 
           (x, a-2, a+2),
           show=  False, 
           legend=True)

p[0].line_color= 'red'
p[1].line_color= 'green'

p[0].label= f'f(x)= ${sm.latex(f)}$'
p[1].label= f'T(x)= ${sm.latex(T)}$'

p.show()
```

```python
#sm.series?
x0= 1
T1= sm.series(f, x=None, x0= x0, n=5, dir='+')
T1= T1.removeO()
T1.expand()
```

```python
p= sm.plot(f, T, T1,
           (x, x0-2, x0+2),
           show=  False, 
           legend=True)

p[0].line_color= 'red'
p[1].line_color= 'green'
p[2].line_color= 'blue'

p[0].label= f'f= ${sm.latex(f)}$'
p[1].label= f'T= ${sm.latex(T)}$'
p[2].label= f'T1= ${sm.latex(T1)}$'

p.show()
```
