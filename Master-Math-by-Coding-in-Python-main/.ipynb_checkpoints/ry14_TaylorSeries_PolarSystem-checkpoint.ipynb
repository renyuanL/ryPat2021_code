```python
## Finding Taylor polynomial

a= 1
f= sm.sin(x)

n= 5

T=   0 #f.subs(x,a) 
for i in range(0, n):
    T += f.diff(x,i).subs(x,a) *(x-a)**i /sm.factorial(i)   
T.expand()
```

```python
## plotting f(x) and T(x) in same window


p= sm.plot(f, T, 
           (x, a-2, a+2),
           show=  False, 
           legend=True)

p[0].line_color= 'red'
p[1].line_color= 'green'

p[0].label= f'f(x)= ${sm.latex(f)}$'
p[1].label= f'T(x)= ${sm.latex(T)}$'

p.show()
```

```python
#sm.series?
x0= 1
T1= sm.series(f, x=None, x0= x0, n=5, dir='+')
T1= T1.removeO()
T1.expand()
```

```python
p= sm.plot(f, T, T1,
           (x, x0-2, x0+2),
           show=  False, 
           legend=True)

p[0].line_color= 'red'
p[1].line_color= 'green'
p[2].line_color= 'blue'

p[0].label= f'f= ${sm.latex(f)}$'
p[1].label= f'T= ${sm.latex(T)}$'
p[2].label= f'T1= ${sm.latex(T1)}$'

p.show()
```

```python
#
# slidable a and b
# 有互動介面
#
%reset -f
%matplotlib qt

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider


plt.subplots_adjust(bottom=0.25)
t = np.arange(0.0, 2*np.pi, 0.001)

a = 1   
b = 1   
r = a + b*np.sin(t)
x = r*np.cos(t)
y = r*np.sin(t)


l, = plt.plot(x,y, lw=4, color='black')
plt.axis([-3, 3, -3, 3])

axcolor = 'lightgoldenrodyellow'
axb  = plt.axes([0.25, 0.10, 0.65, 0.03])  #, axisbg=axcolor)
axa = plt.axes([0.25, 0.15, 0.65, 0.03])   #, axisbg=axcolor)

sa = Slider(axa, 'a', 0, 2, valinit=a)
sb = Slider(axb, 'b', 0, 2, valinit=b)

def update(val):
    a = sa.val     
    b = sb.val   
    
    r = a + b*np.sin(t)
    x = r*np.cos(t)
    y = r*np.sin(t)
    
    l.set_ydata(y)
    l.set_xdata(x)    
    fig.canvas.draw_idle()
    
sa.on_changed(update)
sb.on_changed(update)
plt.title('r = a + b cos(t)')

plt.show()
```
