```python
from ryPat import *
```

```python
def f(x,y):
    z= (x-10)*(x-1)*(x+12)*(y-10)*(y-1)*(y+12)
    return z
x,y= sm.symbols('x,y')
f0= f(x,y).expand()
f0
```

```python
fx= f0.diff(x)
fy= f0.diff(y)
fx
```

```python
fy
```

```python
sols= sm.solve([fx,fy],(x,y))
sols
```

```python
[(x.evalf(), y.evalf()) for (x,y) in sols[-4:]]
```
