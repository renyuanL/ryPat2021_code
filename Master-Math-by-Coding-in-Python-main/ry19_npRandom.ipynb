Notes
-----
The probability density for the Gaussian distribution is

$
p(x) = \frac{1}
            {\sqrt{ 2 \pi \sigma^2 }}
        e^{ - \frac{ (x - \mu)^2 } 
                   {2 \sigma^2} },
$

where 
$\mu$ is the mean and 
$\sigma$  the standard deviation. 

The square of the standard deviation, 
$\sigma^2$,
is called the variance.
