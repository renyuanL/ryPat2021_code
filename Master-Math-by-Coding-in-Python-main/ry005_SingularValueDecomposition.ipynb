# VIDEO: Singular value decomposition
## SVD of Einstein

```python
import numpy as np
import matplotlib.pyplot as plt

# SVD of Einstein
from PIL import Image

#jpgFile= 'ry03.jpg' #
#jpgFile= 'BenLyu.jpg' # 
jpgFile= "BenEinstein.JPG"
jpgFile= "BenEinstein02.JPG"
jpgFile= 'einstein.jpg'

Einstein= Image.open(jpgFile)
Einstein
```

```python
type(Einstein)
```

```python
ein= np.array(Einstein)  # 轉成 np.ndarray 物件
ein.shape  # 仍有3軸，最後1軸是 RGB 顏色軸
```

```python
ein= np.mean(ein, axis=-1) # 沿著最後1軸 取平均 或最大 或 [:,:,0], [:,:,1], [:,:,2]
ein.shape

# 照片或許太大，考慮 down sample
if ein.shape[0]>=256:
    wx= ein.shape[0]//256
    wy= ein.shape[1]//256
    ein= ein[::wx, ::wy]
ein.shape
```

```python
plt.imshow(ein, 'gray')
```

```python
# Take the SVD
U, s, V_T = np.linalg.svd(ein)

Σ= np.diag(s)
V= V_T.T

U, s, V
```

```python
# image the results
fig,ax = plt.subplots(1,3)

ax[0].imshow(U, 'gray')
ax[0].set_title('U')

#ax[1].imshow(np.diag(np.log(s)))
ax[1].imshow(Σ, 'gray')
ax[1].set_title('Σ')

ax[2].imshow(V, 'gray')
ax[2].set_title('V')
plt.show()
```

```python
ein.shape, U.shape,  Σ.shape,  V.shape
```

```python
# 為中間的對角矩陣補0，使成長方形矩陣，可夾在中間讓前後 U, V.T 來乘。

Σ0= np.concatenate((
    Σ,
    np.zeros((U.shape[0]-Σ.shape[0],
              Σ.shape[1]))
    ))
Σ0.shape
```

```python
# 另法補0，似乎比較簡單。

Σ1= np.zeros(ein.shape)
for i in range(len(s)):
    Σ1[i,i]= s[i]
Σ1.shape
```

```python
(Σ0 == Σ1).all()
```

```python
# 原圖重建

Y= U@ Σ0 @V.T

plt.imshow(Y, 'gray')
plt.show()
```

```python
# 重建圖以及原圖
(Y==ein).all()
```

```python
np.allclose(Y,ein)
```

```python
# singular values 也叫做 EigenSpectrum
# plot the eigenspectrum
# 它們的值已經被由大到小的排列。

plt.plot(s,'ks-',markerfacecolor='w')
plt.xlim([-1,50])
plt.xlabel('Component number')
plt.ylabel('Singular value')
plt.title('Eigenspectrum of "Einstein"')
plt.grid()

plt.show()
```

```python
# 更改 sigular values，重建原圖看看。

# reconstruct S from s

print(f'the singular values: shape= {s.shape}')
display(s[0:10])

randorder= np.random.permutation(len(s))

S= np.zeros(np.shape(ein))
for i in range(0,len(s)):
    S[i,i]= s[randorder[i]]  # 把 sigular values 打亂
rein2= U@S@V.T

S= np.zeros(np.shape(ein))
for i in range(0,10):       # 只取 前 10 名 的 singular values
    S[i,i]= s[i]
rein3= U@S@V.T

S= np.zeros(np.shape(ein))
for i in range(0,len(s)):  # 完整取用 所有 314 個 sigular values
    S[i,i]= np.log(s[i])   # 也取了 log
rein4= U@S@V.T

S= np.zeros(np.shape(ein))
for i in range(0,len(s)):  # 完整取用 所有 314 個 sigular values
    S[i,i]= s[i]
rein5= U@S@V.T
```

```python
#--------------------------
plt.figure(figsize=(20,10))

n= 5
plt.subplot(1,n,1)
plt.imshow(ein, 'gray')
plt.axis('off')
plt.title('Original')

plt.subplot(1,n,2)
plt.imshow(rein2, 'gray')
plt.axis('off')
plt.title('rein2')

plt.subplot(1,n,3)
plt.imshow(rein3, 'gray')
plt.axis('off')
plt.title('rein3')

plt.subplot(1,n,4)
plt.imshow(rein4, 'gray')
plt.axis('off')
plt.title('rein4')

plt.subplot(1,n,5)
plt.imshow(rein5, 'gray')
plt.axis('off')
plt.title('rein5')

plt.show()
```
